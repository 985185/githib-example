```python
address = 'Perth, WA, Australia'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

```python
df_comp = pd.read_csv('comp.csv')
df_fc = pd.read_csv('foodcourt.csv')
df_hj = pd.read_csv('Hungry_Jacks.csv')
df_in = pd.read_csv('indian.csv')
df_it = pd.read_csv('italian.csv')
df_kfc = pd.read_csv('KFC.csv')
df_mc = pd.read_csv('Mcdonalds.csv')
df_mx = pd.read_csv('mexican.csv')
df_sb = pd.read_csv('subway.csv')
df_th = pd.read_csv('thai.csv')
df_vt = pd.read_csv('vietnam.csv')
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around the Perth
#popup=folium.Popup(each[1]["GEO_CITY_0"]

# add a red circle marker to represent the Paris
folium.features.CircleMarker( [latitude, longitude],radius=10,color='red', popup='Perth', fill = True, fill_color = 'red', fill_opacity = 0.6).add_to(venues_map)

# add the Mexican restaurants as blue circle markers

for lat, lng, label in zip(df_hj.lat, df_hj.lng, df_hj.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='red',  popup=label,  fill = True, fill_color='red', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_mx.lat, df_mx.lng, df_mx.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='green',  popup=label,  fill = True, fill_color='green', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_mc.lat, df_mc.lng, df_mc.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='blue',  popup=label,  fill = True, fill_color='blue', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_it.lat, df_it.lng, df_it.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='yellow',  popup=label,  fill = True, fill_color='yellow', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_in.lat, df_in.lng, df_in.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='orange',  popup=label,  fill = True, fill_color='orange', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_vt.lat, df_vt.lng, df_vt.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='violet',  popup=label,  fill = True, fill_color='violet', fill_opacity=0.6).add_to(venues_map)
    
for lat, lng, label in zip(df_th.lat, df_th.lng, df_th.categories):
    folium.features.CircleMarker( [lat, lng], radius=5,  color='white',  popup=label,  fill = True, fill_color='white', fill_opacity=0.6).add_to(venues_map)

for lat, lng, label in zip(df_sb.lat, df_sb.lng, df_sb.categories):
    folium.features.CircleMarker( [lat, lng], radius=5, color='black',  popup=label,  fill = True,  fill_color='black', fill_opacity=0.6).add_to(venues_map)
    
for lat, lng, label in zip(df_kfc.lat, df_kfc.lng, df_kfc.categories):
    folium.features.CircleMarker( [lat, lng], radius=5, color='pink', popup=label, fill = True, fill_color='pink', fill_opacity=0.6).add_to(venues_map)
   
for lat, lng, label in zip(df_fc.lat, df_fc.lng, df_fc.categories):
    folium.features.CircleMarker([lat, lng], radius=5, color='coral', popup=label, fill = True, fill_color='coral', fill_opacity=0.6).add_to(venues_map)
    

#for lat, lng, label in zip(df_comp.lat, df_comp.lng, df_comp.categories):
#   folium.features.CircleMarker([lat, lng], radius=5, color='black', popup=label, fill = True, fill_color='orange', fill_opacity=0.6).add_to(venues_map)
# display map
venues_map
```
